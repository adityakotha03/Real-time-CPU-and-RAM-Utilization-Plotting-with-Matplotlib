```python
# Importing the necessary libraries
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from psutil import cpu_percent, virtual_memory
```

```python
# Set the length of the frame to be displayed
frame_len = 200

# Create a new figure object with the given size
fig = plt.figure(figsize=(8, 4))

# Create empty lists to store CPU and RAM utilization data
cpu = []
ram = []

# The animate function that will be called by FuncAnimation
def animate(i):
    # Add current CPU and RAM utilization data to the respective lists
    cpu.append(cpu_percent())
    ram.append(virtual_memory()[2])

    # If the length of the cpu list is less than or equal to the frame_len,
    # plot the entire list; otherwise, plot only the last frame_len data points
    if len(cpu) <= frame_len:
        plt.cla()
        plt.plot(cpu, 'r', label="CPU Utilization")
        plt.plot(ram, 'b', label="RAM Utilization")
    else:
        plt.cla()
        plt.plot(cpu[-frame_len:], 'r', label="CPU Utilization")
        plt.plot(ram[-frame_len:], 'b', label="RAM Utilization")

    # Set the y-axis limits and the labels for the x and y axes
    plt.ylim(0, 100)
    plt.xlabel("Time (s)")
    plt.ylabel("CPU and RAM usage (%)")

    # Add a legend to the plot and adjust the layout
    plt.legend()
    plt.tight_layout()

# Create a FuncAnimation object that will repeatedly call the animate function every 1000ms
ani = FuncAnimation(fig, animate, interval=1000)
```

```python
%matplotlib notebook

frame_len = 200
fig = plt.figure(figsize = (8,4))
cpu = []
ram = []

def animate(i):
    
    cpu.append(cpu_percent())
    ram.append(virtual_memory()[2])
    
    if len(cpu) <= frame_len:
        plt.cla()
        plt.plot(cpu, ls="--", lw=3, marker="X", markersize=8, markerfacecolor="red", markeredgecolor="black", label = "CPU Utilization")
        plt.plot(ram, ls=":", marker="o", markersize=8, markerfacecolor="None", label = "RAM Utilization")
        
    else:
        plt.cla()
        plt.plot(cpu[-frame_len:], ls="--", lw=3, marker="X", markersize=8, markerfacecolor="red", markeredgecolor="black", label = "CPU Utilization")
        plt.plot(ram[-frame_len:], ls=":", marker="o", markersize=8, markerfacecolor="None", label = "RAM Utilization")
        
    plt.ylim(0,100)
    plt.xlabel("Time (s)")
    plt.ylabel("CPU and RAM usage (%)")
    plt.legend()
    plt.tight_layout()
    
    
ani = FuncAnimation(plt.gcf(), animate, interval = 1000)
```
